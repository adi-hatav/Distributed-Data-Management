#### DMA Popularity
For each DMA return a list of all the genres ranked by their popularity in that DMA.

Preprocssing - keeps the relevant columns for this question only.

```python
ref_data_DMA_DID = ref_data.select("dma", "device_id").distinct()
viewing_data_DID_PC = viewing_data.select("device_id","prog_code")
daily_prog_data_PC_G = daily_prog_data.select("prog_code", "genre").distinct()
daily_prog_data_PC_G_splited = daily_prog_data_PC_G.withColumn("genre", explode(split(daily_prog_data_PC_G.genre, ",")))

all_dma = ref_data.select("dma").distinct()
all_genres = daily_prog_data_PC_G_splited.select("genre").distinct()
all_genres_list = [x[0] for x in all_genres.collect()]
all_dma_genre = all_genres.join(all_dma , how="inner")
```

>
First, finds the number of views in a particular genre, for each device.
>
Then, finds the number of views in a particular genre, for each DMA (according to associated devices).
>


```python
genre_per_device = viewing_data_DID_PC.join(daily_prog_data_PC_G_splited , how="inner", on="prog_code")
genre_per_device = genre_per_device.groupBy("device_id", "genre").count().withColumnRenamed("count", "views_per_genre")
DMA_G = genre_per_device.join(ref_data_DMA_DID , how="inner", on="device_id").drop("device_id")

DMA_G1 = DMA_G.groupBy("dma", "genre").sum("views_per_genre").withColumnRenamed("sum(views_per_genre)", "views_per_genre_dma")
DMA_G1.cache()
DMA_G1 = all_dma.join(DMA_G1, how="left", on="dma")
DMA_G1 = DMA_G1.join(all_dma_genre , how="right", on=["dma", "genre"]).na.fill(0).orderBy("dma", "views_per_genre_dma", ascending = 0)
result = DMA_G1.groupBy("dma").agg(collect_list("genre").alias("genre_list")).orderBy("dma")
display(result)
```

####  DMA Popularity according to Wealth
Instead of treating every DMA equally, from now on we consider some DMAs more than others according to their wealth score, determined by their net worth and income.<br>
Where applicable onvert letters to appropriate numbers - A-D = 10-13. Empty/Null values are considered as 0.<br>
Calculate for each DMA their wealth score, which is the normalized average of the net worth and the income: <br>
$\frac{avg\;net\;worth\;in\;dma}{max\;net\;worth\;all\;data} + \frac{avg\;income\;in\;dma}{max\;income\;all\;data}$
For each DMA return its 8 most popular genres - with a limitation - a genre can’t be duplicated.<br>
The wealthiest DMA chooses the top 8 genres according to its popularity rating. After that, the 2nd most wealthy DMA chooses its top 8 genres, from a pool that does not contain the previously taken genres.<br>
We continue like this until there are no genres left - and place empty lists at the remaining DMAs.


```python
dd_HID_IN = demographic_data.select("household_id", "income", "net_worth")#no need to do ditinct

dd_HID_IN = dd_HID_IN.withColumn("update_income", \
                                when((dd_HID_IN.income=='A'), lit(10))\
                                .when((dd_HID_IN.income=='B'), lit(11))\
                                .when((dd_HID_IN.income=='C'), lit(12))\
                                .when((dd_HID_IN.income=='D'), lit(13))\
                                .when((dd_HID_IN.income=="null"), lit(0))\
                                .otherwise(lit(dd_HID_IN.income.cast(IntegerType())))).drop("income")

dd_HID_IN = dd_HID_IN.withColumn("update_net_worth", \
                                when((dd_HID_IN.net_worth=="null"), lit(0))\
                                .otherwise(lit(dd_HID_IN.net_worth.cast(IntegerType())))).drop("net_worth").na.fill(0)

DMA_IM_NW = dd_HID_IN.join(ref_data.select("household_id", "dma"), how="inner", on="household_id").select("update_income", "update_net_worth", "dma")
max_net_worth = dd_HID_IN.select(f.max(dd_HID_IN.update_net_worth)).first()[0]
max_income = dd_HID_IN.select(f.max(dd_HID_IN.update_income)).first()[0]
```

Calculates the "wealth score" for each DMA:

```python
DMA_IM_NW1 = DMA_IM_NW.groupBy("dma").avg("update_income" , "update_net_worth")
DMA_IM_NW1 = DMA_IM_NW1.withColumnRenamed("avg(update_income)", "avg_income")
DMA_IM_NW1 = DMA_IM_NW1.withColumnRenamed("avg(update_net_worth)", "avg_net_worth")
DMA_IM_NW1 = DMA_IM_NW1.withColumn("score", lit((DMA_IM_NW1.avg_net_worth/max_net_worth)+(DMA_IM_NW1.avg_income/max_income)))
DMA_IM_NW1 = DMA_IM_NW1.select("dma", "score")
```

Top 25 DMAs, their wealth score and ordered list of genres according to popularity:

```python
all_dma = ref_data.select("dma").distinct()
all_genres = daily_prog_data_PC_G_splited.select("genre").distinct()
all_genres_list = [x[0] for x in all_genres.collect()]

DMA_score_genre = DMA_IM_NW1.join(result, how="inner", on="dma").orderBy("score",ascending = 0)
DMA_score_genre.cache()
DMA_score_genre1 = DMA_score_genre
dma_and_genre_list =[]
data_to_enter =[] #for Q3
num_of_top_non_empty_dma = 0 #for Q3
for i in range(all_dma.count()):
    curremt_dma = DMA_score_genre.collect()[i].dma 
    current_genres = DMA_score_genre.collect()[i].genre_list
    current_score = DMA_score_genre.collect()[i].score
    str1 =  ""
    for j, genre in enumerate(current_genres):
        if j==8:
            break
        if genre in all_genres_list:
            all_genres_list.remove(genre)
            str1 = str1 + genre + ","
    if str1 != "":
        data_to_enter.append(tuple([curremt_dma, str1[0:-1]]))
        if i < 25:
            num_of_top_non_empty_dma = num_of_top_non_empty_dma + 1
    if i < 25:
        print(DMA_score_genre.collect()[i].dma +", " + f"{DMA_score_genre.collect()[i].score}" + ", " + "[" +str1[0:-1] + "]")
    dma_and_genre_list.append(tuple([curremt_dma, str1[0:-1]]))
schema = StructType([ \
    StructField("dma",StringType(),nullable=True), \
    StructField("genre_list",StringType(), nullable=True)])

q2_res = spark.createDataFrame(data=dma_and_genre_list,schema=schema)
display(q2_res)
```

#### Division into folders by DMA and genre

```python
#create the dataframe that we divide for each site
df = spark.createDataFrame(data=data_to_enter,schema=schema)
df_splited = df.limit(num_of_top_non_empty_dma).withColumn("genre", explode(split(df.genre_list, ","))).drop("genre_list")
daily_prog_splited = daily_prog_data.withColumn("genre", explode(split(daily_prog_data.genre, ",")))
sites = df_splited.join(daily_prog_splited, how="inner", on="genre")
sites.write.option("header", True).partitionBy("dma", "genre").mode("overwrite").csv("dma_genre_partition")
```
